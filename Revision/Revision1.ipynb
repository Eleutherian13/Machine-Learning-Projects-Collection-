## 🎯 Advanced Attributes of `plt.plot()`

Here are the **core categories** of attributes we'll master:

### 1. **Line Appearance**

### 2. **Markers**

### 3. **Colors**

### 4. **Labels and Legends**

### 5. **Axis Control**

## 🔧 1. Line Appearance

```python
plt.plot(x, y, linestyle='--', linewidth=2)
```

* `linestyle` or `ls`:

  * `'solid'` or `'-'`
  * `'dashed'` or `'--'`
  * `'dashdot'` or `'-.'`
  * `'dotted'` or `':'`

* `linewidth` or `lw`: controls thickness

➡️ **Custom Dash Style**:

```python
plt.plot(x, y, linestyle=(0, (5, 10)))  # 5 on, 10 off
```

---

## 🟢 2. Markers

```python
plt.plot(x, y, marker='o', markersize=10, markerfacecolor='red', markeredgewidth=2)
```

* `marker`: `'o'`, `'s'`, `'^'`, `'*'`, `'x'`, `'D'`, etc.
* `markersize` or `ms`: size
* `markerfacecolor` or `mfc`: fill color
* `markeredgecolor` or `mec`: border color
* `markeredgewidth` or `mew`: border thickness

✅ This allows **dot emphasis** on lines for better distinction.

---

## 🎨 3. Colors

You can control color using:

* Named colors: `'red'`, `'green'`, `'blue'`
* Hex codes: `'#FF5733'`
* RGB tuples: `(0.1, 0.2, 0.5)`
* Short codes: `'r'`, `'g'`, `'b'`, `'k'` (black), `'c'` (cyan)

```python
plt.plot(x, y, color='magenta')
```

🔥 Combine line, marker, and color shorthand:

```python
plt.plot(x, y, 'o--r')  # red dashed line with circle markers
```

---

## 🏷️ 4. Labels and Legends

```python
plt.plot(x, y, label="Training Accuracy")
plt.legend(loc='upper left', fontsize=12, title="Legend")
```

* `label`: name of the line
* `legend()`: call to display it
* `loc`: `'upper left'`, `'best'`, `'center'`, etc.
* `title`: title for the legend

---

## 📐 5. Axis Customization

```python
plt.xlim(0, 5)
plt.ylim(0, 40)

plt.xticks([1, 2, 3, 4], ['One', 'Two', 'Three', 'Four'])
plt.xlabel("Epochs")
plt.ylabel("Accuracy (%)")
```

* `xlim`, `ylim`: manual axis limits
* `xticks`, `yticks`: customize tick marks
* `xlabel`, `ylabel`: axis labels
* `title`: graph title

---

## 📊 6. Multiple Lines and Subplots

```python
plt.plot(x, y1, label="Loss", color='red')
plt.plot(x, y2, label="Accuracy", color='blue')
plt.legend()
```

🔀 **Subplots:**

```python
plt.subplot(2, 1, 1)  # 2 rows, 1 column, 1st plot
plt.plot(x, y1)

plt.subplot(2, 1, 2)
plt.plot(x, y2)
```

---

## ⚙️ 7. Precision & Anti-aliasing

```python
plt.plot(x, y, antialiased=True)  # smooths the curve
```

Other fine-tuning:

* `dash_capstyle='round'`: makes dashes round-ended
* `solid_joinstyle='miter'`: changes line joining angles

---

## 🧠 8. Interactive Notebooks (Jupyter)

```python
%matplotlib inline   # static plots
%matplotlib notebook # interactive plots (zoom/pan)
```

Also:

* `plt.grid(True)` adds a grid
* `plt.tight_layout()` adjusts spacing

## 🔥 Pro Example (All in One)

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [2, 3, 5, 10]

plt.figure(figsize=(8, 5))

plt.plot(x, y1, 'o--b', label="Model A", linewidth=2, markersize=8)
plt.plot(x, y2, 's-.r', label="Model B", linewidth=2, markersize=8)

plt.xlabel("Epochs", fontsize=12)
plt.ylabel("Performance", fontsize=12)
plt.title("Model Comparison", fontsize=14)

plt.legend(loc='upper left', title="Models")
plt.grid(True)
plt.xlim(0, 5)
plt.ylim(0, 20)

plt.tight_layout()
plt.show()
```

````python
import matplotlib.pyplot as plt

# Sample data
years = [2021, 2018, 2022, 2023, 2017]
category1 = [20, 25, 30, 28, 35]
category2 = [10, 15, 20, 18, 25]
category3 = [5, 8, 10, 9, 15]
plt.figure(figsize=(10, 6) , frameon=True , facecolor='beige' , edgecolor='black')
# Create a area plot
plt.stackplot(years, category1, category2, category3, labels=['Category 1', 'Category 2', 'Category 3'], colors=['magenta', 'salmon', 'lightgreen'] , edgecolor='black' , linewidth =3 , linestyle = '-' , hatch = '//') 
xticks = [2017, 2018, 2019, 2020, 2021, 2022, 2023]
plt.xticks(ticks = xticks , rotation = 45)
# Add labels and title
plt.xlabel("Years" , fontdict={'family' : 'serif' , 'color' : 'brown' , 'weight' : 'bold' , 'size' : 12} , fontsize = 16)
plt.ylabel("Values" , fontdict={'family' : 'serif' , 'color' : 'brown' , 'weight' : 'bold' , 'size' : 12} , fontsize = 16)
plt.title("Area Plot" , fontsize = 18 , fontweight = 'bold' , color = 'brown')
plt.legend(fontsize = 12 , loc = 'upper right')

plt.show()

````
